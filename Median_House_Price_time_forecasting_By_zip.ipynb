#### Yearly Maintience Cost 

https://porch.com/resource/cost-of-home-maintenance-in-america

The task was to find the increase in value of property after 1 year, working under this assumption: the goal is to see how much money can be made from the purchase and holding of property for 1 year.  The cost of maintaining that property should be taken into account.  Instead of subtracting the cost of maintience from the potential profit the state average mainteince cost difference from the average of the entire country will be applied to each zipcode.  This means that if a state's average mainteince cost is higher than the national average ($16,133) the potential gain will be reduced by that much.  if the cost is less the difference will be added (subtracting a negative) because the investment would be less.  

```python
# Average cost to maintain a home per state placed into a dictionary.
state_maintain_cost =  {'NJ':18151,'DC':18119,'CT':17782,'MD':17699,'MA':17461,
                       'RI':17034,'CA':16957,'NH':16953,'DE':16941,'NY':16697,
                       'HI':16679,'NV':16498,'WY':16478,'CO':16432,'VA':16422,
                       'IL':16315,'WA':16304,'VT':16246,'MN':16228,'UT':16227,
                       'AZ':16097,'FL':16095,'PA':16082,'WI':16065,'AK':16061,
                       'IA':16017,'IN':15959,'OH':15932,'NE':15919,'TX':15917,
                       'KS':15838,'ND':15819,'ID':15772,'OR':15754,'MI':15731,
                       'GA':15659,'NC':15658,'SD':15649,'MO':15638,'ME':15593,
                       'MT':15507,'OK':15430,'LA':15429,'AL':15367,'NM':15354,
                       'TN':15335,'SC':15288,'KY':15129,'AR':15115,'WV':15068,'MS':14890}
```

```python
#Finding the national average of maintaining a home in the USA.
cost=0
i=0
for key,value in state_maintain_cost.items():
    cost=cost+value
    i=i+1
avg_cost=cost/(i)
```

```python
# Average Cost to maintain a home.
avg_cost
```

```python
# Retrieving the cost to maintain a home in each zip based on state.
main_cost=[]
for statename in state:
    main_cost.append(state_maintain_cost.get(statename))
main_cost[0:11]
```

```python
#changing this number to be the difference from the national average.
avg_main_cost=[]
for cost in main_cost:
    avg_main_cost.append(cost-avg_cost)
```

```python
avg_main_cost[0:11]
```

By looking at the 11th zipcode it can be seen that the cost of maintaining a property in that zip code will be en extra investment of $2017.86 for the year which should be factored into potential profits made.

# Function Creation

In order to model each zip code 3 different functions were created.  The fucntions obtain the data for one specific zipcode then split the data into training (data before 2018) and testing (data from 2018).  The data for the zipcodes are then differenced using the ARIMA model creation each different function attempts a different differencing.  The predictions are saved first being transfromed to string then to float which allows them to be saved into a data frame.

```python
from statsmodels.tsa.arima.model import ARIMA
import numpy as np
```

```python
def zip_pull_train(num):
     
    tempdf=df[df['ZIP']==num]
    tempdf=tempdf.reset_index()
    temp_train,temp_test=tempdf[0:264],tempdf[264:276]
    x=temp_train['Home_value'].to_list()
    temp_test=temp_test.reset_index()
    y=temp_test['Home_value'].to_list()
                
    history = [x for x in x]
    predictions = []
    # walk-forward validation
    for t in range(len(y)):
        model = ARIMA(history, order=(5,1,2))
        model_fit = model.fit()
        output = model_fit.forecast()
        yhat = output[0]
        predictions.append(yhat)
        obs = y[t]
        history.append(obs)
     
    temp2=[]
    for item in predictions:
        temp2.append(str(item))

    temp3=[]
    for item in temp2:
        temp3.append(float(item))
        
    return temp3
```

```python
def zip_pull_train_0(num):
     
    tempdf=df[df['ZIP']==num]
    tempdf=tempdf.reset_index()
    temp_train,temp_test=tempdf[0:264],tempdf[264:276]
    x=temp_train['Home_value'].to_list()
    temp_test=temp_test.reset_index()
    y=temp_test['Home_value'].to_list()
            
    history = [x for x in x]
    predictions = []
    # walk-forward validation
    for t in range(len(y)):
        model = ARIMA(history, order=(5,0,2))
        model_fit = model.fit()
        output = model_fit.forecast()
        yhat = output[0]
        predictions.append(yhat)
        obs = y[t]
        history.append(obs)
     
    temp2=[]
    for item in predictions:
        temp2.append(str(item))

    temp3=[]
    for item in temp2:
        temp3.append(float(item))
        
    return temp3
```

```python
def zip_pull_train_2(num):
     
    tempdf=df[df['ZIP']==num]
    tempdf=tempdf.reset_index()
    temp_train,temp_test=tempdf[0:264],tempdf[264:276]
    x=temp_train['Home_value'].to_list()
    x_log=np.log(x)
    temp_test=temp_test.reset_index()
    y=temp_test['Home_value'].to_list()
            
    history = [x for x in x]
    predictions = []
    # walk-forward validation
    for t in range(len(y)):
        model = ARIMA(history, order=(5,2,2))
        model_fit = model.fit()
        output = model_fit.forecast()
        yhat = output[0]
        predictions.append(yhat)
        obs = y[t]
        history.append(obs)
        
    temp2=[]
    for item in predictions:
        temp2.append(str(item))

    temp3=[]
    for item in temp2:
        temp3.append(float(item))
        
    return temp3
```
